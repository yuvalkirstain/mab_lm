```python
import pandas as pd
```

```python
df = pd.read_json("../raw_data/wiki.valid.tokens.title_map.json", orient="index")
df.head(3)
```

```python
df.what.value_counts()
```

```python
df_cats = df[['query_title', 'title', 'categories']].explode('categories').reset_index()
df_cats.categories.value_counts()
```

```python
df_labels = df[['query_title', 'title', 'labels']]
df_labels["labels_list"] = df_labels.labels.apply(lambda dic: [v for k, v in dic.items()] if type(dic) == dict else [])
df_labels = df_labels.explode('labels_list').reset_index()
df_labels.labels_list.value_counts()
```
